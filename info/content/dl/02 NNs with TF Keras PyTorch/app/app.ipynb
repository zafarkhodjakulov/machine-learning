```python
import tensorflow as tf

model = tf.keras.models.load_model('fashion_mnist_clf.keras')
```

```python
model
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boot']
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

with Image.open('images/ankle_boot1.jpg') as image:
    display(image)
    image = image.convert(mode='F')
    image = image.resize((28, 28))
    img_arr = np.array(image)
    img_arr = img_arr / 255
```

```python
img_arr.shape
```

```python
y_pred = model.predict(np.array([img_arr]))
```

```python
class_names[y_pred.argmax()]
```

```python
plt.imshow(img_arr, cmap='binary')
plt.axis('off');
```
