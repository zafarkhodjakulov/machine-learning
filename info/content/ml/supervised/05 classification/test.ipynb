```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False)
```

```python
X = mnist.data
X.shape
```

```python
y = (mnist.target == '5').astype(int)
y
```

```python
X_tr, X_te, y_tr, y_te = X[:60000], X[60000:], y[:60000], y[60000:]
X_tr = X_tr / 255.0
X_te = X_te / 255.0
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_tr, y_tr)
```

```python
log_reg.score(X_tr, y_tr)
```

```python
log_reg.score(X_te, y_te)
```

```python
pd.DataFrame(data=y, columns=['y']).value_counts()
```

```python
class DummyModel:
    def fit(self, X, y):
        return self
    
    def predict(self, X):
        n = X.shape[0]
        return np.zeros(n)
    
    def score(self, X, y):
        y_pred = self.predict(X)
        return accuracy_score(y, y_pred)
```

```python
clf = DummyModel()
clf.fit(X_tr, y_tr)
```

```python
clf.score(X_tr, y_tr)
```

```python
clf.score(X_te, y_te)
```

```python
y_pred = clf.predict(X_tr)
precision_score(y_tr, y_pred)
```
