Linear Regression is a key data science tool for predicting continuous outcomes. Linear regression predicts the relationship between two variables by assuming they have a straight-line connection. It finds the best line that minimizes the differences between predicted and actual values. Linear regression is a quiet and the simplest statistical regression technique used for predictive analysis in machine learning.

### Key Concepts: Dependent and Independent Variables

In linear regression, understanding the roles of **dependent** and **independent variables** is essential. These variables form the basis of the predictive relationship that linear regression models. Let’s dive into each concept and how they interact in regression.

**What is a Dependent Variable?**
The **dependent variable** (also known as the response variable) is the primary outcome or the value we aim to predict or explain. It’s called "dependent" because its value depends on the changes in another variable – the independent variable. 

In a regression equation, the dependent variable is often denoted by **$Y$**. For example:
- In predicting **sales** based on advertising spend, **sales** would be the dependent variable.
- In estimating **weight** based on height, **weight** is the dependent variable.
  
The dependent variable is the target of our analysis, representing the aspect of the data we’re most interested in understanding or predicting.

**What is an Independent Variable?**
The **independent variable** (also called the predictor or explanatory variable) is the factor that we believe influences or explains the dependent variable. The independent variable is "independent" because we assume its values are not influenced by changes in the dependent variable.

In a regression equation, the independent variable is often denoted by **$X$**. For example:
- When predicting sales, **advertising spend** could be the independent variable, as we expect it to influence sales.
- In estimating weight based on height, **height** would be the independent variable, as it likely affects weight.

The independent variable is the input of our model, and we use it to make predictions about the dependent variable.

**Relationship Between Dependent and Independent Variables**
In simple linear regression, we model the relationship between a single independent variable ($X$) and a single dependent variable ($Y$) with a straight line. The goal is to determine the linear relationship between these variables, which can be represented by the equation:

$$
Y = mX + b + \varepsilon
$$

Where:
- **$Y$** is the predicted value of the dependent variable.
- **$X$** is the value of the independent variable.
- **$m$** (slope) represents the rate of change in $Y$ for every unit change in $X$.
- **$b$** (intercept) is the starting value of $Y$ when $X$ is zero.
- **$\varepsilon$** (error term) is the random noise or residual.


**Examples of Dependent and Independent Variables**
To clarify the distinction, here are some examples from various fields:

- **Business**: Predicting **revenue** (dependent variable) based on **marketing budget** (independent variable).
- **Healthcare**: Estimating **blood pressure** (dependent variable) based on **age** (independent variable).
- **Environmental Science**: Measuring **carbon emissions** (dependent variable) based on **industrial activity** (independent variable).
- **Education**: Analyzing **test scores** (dependent variable) based on **study hours** (independent variable).

By correctly identifying the dependent and independent variables, we ensure that our regression analysis is accurate and meaningful. Understanding these roles helps us set up a model that can uncover relationships and make reliable predictions.

### Line of Best Fit and the Linear Relationship

In linear regression, we aim to find the **line of best fit**—a straight line that best represents the relationship between an independent variable $X$ and a dependent variable $Y$. This line shows the **linear relationship** between these two variables, meaning that changes in $X$ correspond to consistent changes in $Y$. Let’s break down what the line of best fit is, how it’s calculated, and why it’s useful.

Mathematically, the line of best fit in simple linear regression can be represented by the equation:

$$
\hat{Y} = mX + b
$$

Where:
- **$\hat{Y}$** is the predicted value of the dependent variable.
- **$m$** is the slope, indicating how much $\hat{Y}$ changes for each unit increase in $X$.
- **$X$** is the independent variable.
- **$b$** is the intercept, which is the value of $\hat{Y}$ when $X = 0$.

#### Calculating the Line of Best Fit
To find the line of best fit, we calculate the slope ($m$) and intercept ($b$) values that minimize the error (or residuals). This process is usually done using a method called **Ordinary Least Squares (OLS)**. The goal of OLS is to find the values of $m$ and $b$ that result in the smallest possible sum of squared residuals.

In practical terms, most data analysis tools (such as Python’s `scikit-learn`, R, or Excel) can compute the line of best fit automatically.

#### Interpreting the Slope and Intercept
- **Slope ($m$)**: The slope represents the rate at which the dependent variable $Y$ changes with each unit increase in the independent variable $X$. A positive slope indicates a positive relationship (as $X$ increases, $Y$ also increases), while a negative slope indicates a negative relationship (as $X$ increases, $Y$ decreases).
  
- **Intercept ($b$)**: The intercept is the point where the regression line crosses the $Y$-axis. It represents the predicted value of $Y$ when $X = 0$. In some cases, this has a meaningful interpretation, but in others (like predicting sales based on ad spending), it may not always make practical sense.

#### Importance of the Linear Relationship
A **linear relationship** means that the change in the dependent variable $Y$ is proportional to the change in the independent variable $X$. In a strong linear relationship, the data points in a scatter plot tend to align closely along the line of best fit, while a weaker linear relationship shows more scattered points around the line.

- **Positive Linear Relationship**: As $X$ increases, $Y$ also increases. For example, if we observe that higher temperatures ($X$) correlate with higher ice cream sales ($Y$), we have a positive relationship.
  
- **Negative Linear Relationship**: As $X$ increases, $Y$ decreases. For instance, as the number of miles driven ($X$) increases, the remaining gas in a tank ($Y$) decreases.

### Correlation

In linear regression, understanding **correlation** is essential because it tells us about the strength and direction of the relationship between two variables. **Correlation** measures how closely two variables move in relation to each other. When variables are correlated, changes in one variable tend to be associated with changes in another, which helps determine whether a linear model is appropriate.

#### What is the Correlation Coefficient?
The **correlation coefficient** (often denoted as **$( r )$**) is a numerical value that quantifies the strength and direction of the relationship between two variables. The correlation coefficient ranges from **-1 to +1**:
- **$( r = +1 )$** indicates a perfect positive linear relationship.
- **$( r = -1 )$** indicates a perfect negative linear relationship.
- **$( r = 0 )$** suggests no linear relationship between the variables.

In general:
- **Positive Correlation**: When $( r )$ is positive, it means that as one variable increases, the other also increases. For example, there is typically a positive correlation between advertising spend and sales.
- **Negative Correlation**: When $( r )$ is negative, it indicates that as one variable increases, the other decreases. For instance, there’s often a negative correlation between hours of exercise and body weight.
- **No Correlation**: If $( r )$ is close to zero, there is no consistent pattern of increase or decrease between the two variables.

#### How to Interpret the Correlation Coefficient
The closer the value of $( r )$ is to **+1** or **-1**, the stronger the linear relationship between the variables. Here’s a general guide:
- **$( r )$ between 0.7 and 1.0** (or -0.7 and -1.0): Strong correlation
- **$( r )$ between 0.3 and 0.7** (or -0.3 and -0.7): Moderate correlation
- **$( r )$ between 0.0 and 0.3** (or -0.0 and -0.3): Weak or no correlation

#### Calculating the Correlation Coefficient
The correlation coefficient $( r )$ is calculated using the formula:

$$
r = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum (X_i - \bar{X})^2 \cdot \sum (Y_i - \bar{Y})^2}}
$$


Where:
- $( X )$ and $( Y )$ are the variables being correlated.
- $( \bar{X} )$ and $( \bar{Y} )$ are the means of $( X )$ and $( Y )$.
- The numerator represents the co-variation between $( X )$ and $( Y )$, and the denominator adjusts for the overall variability in $( X )$ and $( Y )$.

Most data analysis tools calculate $( r )$ automatically, but understanding the formula helps in interpreting the concept.

#### Correlation vs. Causation
It’s important to remember that **correlation does not imply causation**. Just because two variables have a high correlation does not mean that one variable causes the other to change. Correlation simply shows that there’s an association between the two variables, not that one variable is the direct cause of changes in the other.

For example, there may be a strong correlation between ice cream sales and drowning incidents, but this does not mean that eating ice cream causes drowning. The underlying cause for both may be an external factor, like hotter weather.

#### Examples of Correlation in Real Life
1. **Education and Income**: There is often a positive correlation between years of education and income level. As education increases, income tends to increase as well.
  
2. **Temperature and Energy Consumption**: In colder months, there may be a negative correlation between outdoor temperature and energy consumption, as people use more heating when temperatures drop.

3. **Experience and Job Performance**: There may be a positive correlation between years of experience and job performance, as experience often enhances skills and productivity.

#### Why Correlation Matters in Linear Regression
Correlation is a key concept in linear regression because it tells us whether it’s appropriate to use a linear model. If the correlation between two variables is strong, a linear regression model can often provide reliable predictions. On the other hand, if the correlation is weak, a linear model may not be suitable for the data.

### Measuring Data Variability

In linear regression, it's crucial not only to understand the relationship between variables but also to measure how much data fluctuates or deviates from the expected values. This is where **data variability** comes into play. Understanding and measuring variability helps to assess the reliability and precision of your model's predictions. Two key metrics for measuring variability are **standard deviation** and **variance**.

#### What is Variability?
**Variability** refers to the spread or dispersion of data points in a dataset. In simpler terms, it tells us how much the data points differ from one another. When the data points are spread out, variability is high; when they are clustered closely together, variability is low.

In the context of linear regression, variability is important because it indicates how well the regression line fits the data. A model with low variability means that the predictions are more precise, and the data points are close to the line of best fit. High variability suggests that the data points are widely scattered, meaning the model may not explain the relationship between the variables well.

#### Variance
Variance is a statistical measure that describes the degree of spread in the data. It is the average of the squared differences between each data point and the mean of the dataset. In simple terms, variance measures how far data points are from the mean value and each other.

The formula for variance $( \sigma^2 )$ for a population is:

$$
\sigma^2 = \frac{\sum (X_i - \mu)^2}{N}
$$

Where:
- $( X_i )$ represents each data point.
- $( \mu )$ is the mean of the data.
- $( N )$ is the total number of data points.

For a sample (rather than the whole population), the formula becomes:

$$
s^2 = \frac{\sum (X_i - \bar{X})^2}{n-1}
$$

Where:
- $( \bar{X} )$ is the sample mean.
- $( n )$ is the sample size.

Variance is a critical measure because it quantifies the spread of the data. However, the units of variance are the square of the original units, which makes it harder to interpret directly.

#### Standard Deviation
**Standard deviation** is the square root of variance and is a more commonly used measure of variability. Unlike variance, the standard deviation has the same units as the original data, making it easier to understand in real-world terms.

The formula for the standard deviation $( \sigma )$ for a population is:

$$
\sigma = \sqrt{\frac{\sum (X_i - \mu)^2}{N}}
$$

For a sample, it’s calculated as:

$$
s = \sqrt{\frac{\sum (X_i - \bar{X})^2}{n-1}}
$$

Standard deviation tells us, on average, how far each data point is from the mean of the dataset. A smaller standard deviation indicates that the data points are clustered close to the mean, while a larger standard deviation suggests that the data points are spread out.

#### Importance of Measuring Data Variability in Regression
In the context of linear regression, measuring variability helps assess how well the regression model fits the data:
- **Low Variability**: When the variability is low, the data points are close to the regression line, meaning the model does a good job of predicting the dependent variable.
- **High Variability**: If the variability is high, the data points are far from the regression line, which indicates that the model may not be capturing the underlying relationship well. This might require further exploration, such as considering additional variables, transforming the data, or using a more complex model.

#### Visualizing Data Variability
In a scatter plot, you can visualize variability by looking at how spread out the data points are. If the data points are tightly clustered around the regression line, the variability is low, and the model is likely a good fit. On the other hand, if the data points are spread out widely, the model may not be capturing the relationship well.

#### Example
Let’s say you’re predicting **sales** based on **ad spend** in a simple linear regression model. If you find that the **standard deviation of sales** is large compared to the mean sales value, it suggests that sales vary widely, and the model might have a harder time predicting outcomes accurately. Conversely, if the standard deviation is small, the model’s predictions will likely be more reliable.

### Building a Simple Linear Regression Model

In this section, we will walk through the process of building a simple linear regression model using the `advertising.csv` dataset. This dataset contains information about the amount of money spent on different types of advertising (TV, radio, and newspaper) and the corresponding sales of a product. We will use this data to predict sales based on TV advertising spend, demonstrating the steps involved in creating and evaluating a linear regression model.

#### 1. **Importing Libraries and Loading Data**
First, let’s import the necessary libraries and load the `advertising.csv` dataset into a pandas DataFrame.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
data = pd.read_csv('data/advertising.csv')

# Display the first few rows of the dataset
data.head()
````

#### 2. **Exploring the Dataset**
The `advertising.csv` dataset contains the following columns:
- **TV**: Amount spent on TV advertising (in thousands of dollars)
- **Radio**: Amount spent on radio advertising (in thousands of dollars)
- **Newspaper**: Amount spent on newspaper advertising (in thousands of dollars)
- **Sales**: Sales of the product (in thousands of units)

Let’s begin by exploring the dataset visually to understand the relationship between the variables.

````python
# Visualizing the relationship between TV advertising and sales
plt.figure(figsize=(8, 5))
sns.scatterplot(x='TV', y='Sales', data=data)
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Advertising Spend (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.show()
````

This scatter plot will give us a sense of the relationship between TV advertising spend and sales.

#### 3. **Preparing the Data**
For this example, we will focus on predicting **Sales** based on **TV** advertising spend. This means that:
- The **independent variable (X)** is the **TV** column.
- The **dependent variable (Y)** is the **Sales** column.

We need to split the data into training and testing sets to evaluate the model’s performance.

````python
# Define the independent (X) and dependent (Y) variables
X = data[['TV']]  # Independent variable (TV spend)
Y = data['Sales']  # Dependent variable (Sales)

# Split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                            test_size=0.2, random_state=42)
````

#### 4. **Building the Linear Regression Model**
Now, let’s create a simple linear regression model using the training data and fit it to the data.

````python
# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, Y_train)
````

#### 5. **Making Predictions**
Once the model is trained, we can use it to predict sales based on TV advertising spend in the test set.

````python
# Make predictions using the test set
Y_pred = model.predict(X_test)

# Plot the regression line along with the data points
plt.figure(figsize=(8, 5))
plt.scatter(X_test, Y_test, color='blue', label='Actual Sales')
plt.plot(X_test, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression: TV Advertising vs Sales')
plt.xlabel('TV Advertising Spend (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.legend()
plt.show()
````

This plot will show the actual sales as blue points and the predicted sales (based on the regression model) as a red line.

#### 6. **Evaluating the Model**
To assess how well the model performs, we can calculate performance metrics such as the **Mean Squared Error (MSE)** and **R-squared (R²)**.

````python
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y_test, Y_pred)
print(f'Mean Squared Error (MSE): {mse}')

# Calculate R-squared (R²)
r2 = r2_score(Y_test, Y_pred)
print(f'R-squared (R²): {r2}')
````

- **Mean Squared Error (MSE)**: This measures the average of the squared differences between actual and predicted values. Lower values of MSE indicate a better fit.
- **R-squared (R²)**: This measures how well the independent variable explains the variability of the dependent variable. An R² value closer to 1 indicates a better fit.

#### 7. **Interpreting the Results**
- **Regression Coefficient**: The `model.coef_` attribute gives the slope of the regression line, which represents how much **sales** change for every unit increase in **TV advertising spend**.

````python
# Get the model coefficient (slope) and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f'Regression Coefficient (Slope): {slope}')
print(f'Intercept: {intercept}')
````

- **Equation of the Line**: The equation of the regression line is:
  $$
  \text{Sales} = (\text{Slope}) \times (\text{TV Advertising Spend}) + \text{Intercept}
  $$
  For example, if the slope is 0.045 and the intercept is 5, the equation would be:
  $$
  \text{Sales} = 0.045 \times (\text{TV Advertising Spend}) + 5
  $$

#### 8. **Conclusion**
In this example, we built a simple linear regression model to predict **Sales** based on **TV** advertising spend. By using the model, we learned how to:
- Load and explore a dataset
- Split data into training and testing sets
- Train a linear regression model
- Make predictions and evaluate the model’s performance

While this model is simple, it demonstrates the foundational steps involved in building a linear regression model. You can extend this to more complex models by adding additional independent variables, such as **radio** and **newspaper** advertising, to improve prediction accuracy.

<h2 style="text-align:center;color:#0F4C81">Solving simple linear regression with pure Python</h2>

````python
def plot_data(x, y, y_pred = None, coefs = None):
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, alpha=0.8, s=10, c='b', label='data points')
    if y_pred is not None:
        plt.plot(x, y_pred, color='red',
                 label=fr"$y_i={coefs['slope']:.2f} + {coefs['coef']:.2f}\cdot x_i$",)
    plt.xlabel('hours')
    plt.ylabel('scores')
    plt.legend()
    plt.show()
````

Dataset: `data/student_scores.csv`

````python
import csv
import matplotlib.pyplot as plt

data_file = 'data/student_scores.csv'

with open(data_file) as f:
    reader = csv.reader(f)
    columns = next(reader)
    data = [[float(val) for val in row] for row in reader]
    
# print(columns)
# print(data)

hours = [val[0] for val in data]
scores = [val[1] for val in data]

plot_data(hours, scores)
````

````python
import statistics

x = hours.copy()
y = scores.copy()

x_mean = statistics.mean(x)
y_mean = statistics.mean(y)
xy_cov = statistics.covariance(x, y)
x_var = statistics.variance(x)

coef = xy_cov / x_var
slope = y_mean - coef * x_mean

f = lambda x: slope + coef * x

y_pred = [f(x_i) for x_i in x]

plot_data(x, y, y_pred, {'coef': coef, 'slope': slope})
````

<h2 style="text-align:center;color:#0F4C81">Solving simple linear regression with NumPy</h2>

````python
import numpy as np
import matplotlib.pyplot as plt

data_file = 'data/student_scores.csv'
data = np.loadtxt(data_file, delimiter=',', skiprows=1)

x = data[:, 0]
y = data[:, 1]

x_mean = np.mean(x)
y_mean = np.mean(y)

numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)

coef = numerator / denominator
slope = y_mean - coef * x_mean

y_pred = slope + coef * x

plot_data(x, y, y_pred, {'coef': coef, 'slope': slope})
````

## More examples

To find the linear regression coefficients (including the intercept) for this data without using any libraries, we can use the **Normal Equation** method:

$$
\beta = (X^T X)^{-1} X^T y
$$

### Steps to Solve in Pure Python:
1. **Augment the Input Matrix \( X \)** to include a column of 1’s for the intercept.
2. **Calculate $ X^T X $**.
3. **Calculate $ (X^T X)^{-1} $**, the inverse of $ X^T X $.
4. **Calculate $ X^T y $**.
5. Multiply the resulting matrices to get $ \beta $, which contains both the intercept and the coefficients.

Here’s the pure Python code to achieve this:

```python
# Data
X = [
    [230.1, 37.8],
    [44.5, 39.3],
    [17.2, 45.9],
    [151.5, 41.3],
    [180.8, 10.8],
    [8.7, 48.9],
    [57.5, 32.8],
    [120.2, 19.6],
    [8.6, 2.1],
    [199.8, 2.6]
]
y = [22.1, 10.4, 12.0, 16.5, 17.9, 7.2, 11.8, 13.2, 4.8, 15.6]

# Step 1: Augment X with a column of 1's for the intercept
X_augmented = [[1] + row for row in X]

# Helper function to perform matrix multiplication
def matrix_multiply(A, B):
    result = [[sum(a * b for a, b in zip(A_row, B_col)) for B_col in zip(*B)] for A_row in A]
    return result

# Helper function to calculate transpose of a matrix
def transpose(matrix):
    return list(map(list, zip(*matrix)))

# Helper function to calculate the inverse of a 3x3 matrix
def inverse_3x3(matrix):
    det = (
        matrix[0][0] * (matrix[1][1] * matrix[2][2] - matrix[1][2] * matrix[2][1]) -
        matrix[0][1] * (matrix[1][0] * matrix[2][2] - matrix[1][2] * matrix[2][0]) +
        matrix[0][2] * (matrix[1][0] * matrix[2][1] - matrix[1][1] * matrix[2][0])
    )
    inv_det = 1 / det
    inverse = [
        [
            inv_det * (matrix[1][1] * matrix[2][2] - matrix[1][2] * matrix[2][1]),
            inv_det * (matrix[0][2] * matrix[2][1] - matrix[0][1] * matrix[2][2]),
            inv_det * (matrix[0][1] * matrix[1][2] - matrix[0][2] * matrix[1][1]),
        ],
        [
            inv_det * (matrix[1][2] * matrix[2][0] - matrix[1][0] * matrix[2][2]),
            inv_det * (matrix[0][0] * matrix[2][2] - matrix[0][2] * matrix[2][0]),
            inv_det * (matrix[0][2] * matrix[1][0] - matrix[0][0] * matrix[1][2]),
        ],
        [
            inv_det * (matrix[1][0] * matrix[2][1] - matrix[1][1] * matrix[2][0]),
            inv_det * (matrix[0][1] * matrix[2][0] - matrix[0][0] * matrix[2][1]),
            inv_det * (matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]),
        ],
    ]
    return inverse

# Step 2: Calculate X^T * X
X_T = transpose(X_augmented)
X_T_X = matrix_multiply(X_T, X_augmented)

# Step 3: Calculate (X^T * X)^-1
X_T_X_inv = inverse_3x3(X_T_X)

# Step 4: Calculate X^T * y
y_as_matrix = [[val] for val in y]
X_T_y = matrix_multiply(X_T, y_as_matrix)

# Step 5: Calculate coefficients (X^T * X)^-1 * (X^T * y)
beta = matrix_multiply(X_T_X_inv, X_T_y)

# Extract coefficients
intercept = beta[0][0]
coefficients = [beta[i][0] for i in range(1, len(beta))]

print("Intercept:", intercept)
print("Coefficients:", coefficients)
```

### Explanation of the Output:
- **Intercept**: The constant term in the regression equation.
- **Coefficients**: The slopes associated with each feature in $ X $.

Using `numpy` will simplify and streamline the calculations. With `numpy`, we can directly use matrix operations like transposition, multiplication, and inversion. Here’s how to compute the linear regression coefficients (including the intercept) with `numpy`:

```python
import numpy as np

# Data
X = np.array([
    [230.1, 37.8],
    [44.5, 39.3],
    [17.2, 45.9],
    [151.5, 41.3],
    [180.8, 10.8],
    [8.7, 48.9],
    [57.5, 32.8],
    [120.2, 19.6],
    [8.6, 2.1],
    [199.8, 2.6]
])
y = np.array([22.1, 10.4, 12.0, 16.5, 17.9, 7.2, 11.8, 13.2, 4.8, 15.6])

# Step 1: Add a column of ones to X for the intercept term
X_augmented = np.c_[np.ones(X.shape[0]), X]

# Step 2: Calculate the coefficients using the Normal Equation
# beta = (X.T * X)^(-1) * X.T * y
beta = np.linalg.inv(X_augmented.T @ X_augmented) @ X_augmented.T @ y

# Extract intercept and coefficients
intercept = beta[0]
coefficients = beta[1:]

print("Intercept:", intercept)
print("Coefficients:", coefficients)
```

### Explanation
1. **Adding the intercept**: `np.c_[np.ones(X.shape[0]), X]` adds a column of ones to the input matrix $ X $ to account for the intercept.
2. **Normal Equation**: We compute $ \beta = (X^T X)^{-1} X^T y $ using `@` for matrix multiplication.
3. **Output**: `intercept` is the first value in `beta`, and the remaining values are the coefficients.

Using `statsmodels`, we can calculate the linear regression coefficients easily with a high-level API that provides additional statistical details about the model. Here’s how to do it:

```python
import numpy as np

# Step 1: Add a constant (intercept) to the input data
X_augmented = sm.add_constant(X)

# Step 2: Fit the model
model = sm.OLS(y, X_augmented)
results = model.fit()

# Display the results
print("Intercept:", results.params[0])
print("Coefficients:", results.params[1:])
print("\nSummary:\n", results.summary())
```

### Explanation
1. **Adding the Intercept**: `sm.add_constant(X)` adds a column of ones to `X` to handle the intercept term automatically.
2. **Fitting the Model**: `sm.OLS(y, X_augmented)` initializes the Ordinary Least Squares regression model, and `results = model.fit()` fits the model to the data.
3. **Results**: `results.params[0]` is the intercept, and `results.params[1:]` gives the coefficients for each feature in `X`. The `results.summary()` provides a comprehensive summary, including R-squared, p-values, and other useful statistics.

With `scikit-learn`, calculating the linear regression coefficients is straightforward. The `LinearRegression` class handles the model fitting and provides methods to retrieve the intercept and coefficients. Here’s how to do it:

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([
    [230.1, 37.8],
```

In addition to `statsmodels` and `scikit-learn`, there are several other libraries and tools available in Python that can help with linear regression tasks. Here are some of the notable ones:

### 1. **TensorFlow / Keras**
   - **Use case**: While primarily used for deep learning, you can use TensorFlow (or Keras, which is built on top of TensorFlow) for simple linear regression as well. It provides automatic differentiation and optimization capabilities for training the model.
   - **How to use**: You can define a simple linear model using `tf.keras.Sequential` and use the built-in training loops or the `tf.optimizers` to minimize the loss function (e.g., Mean Squared Error).
   - **Example**:
     ```python
     import tensorflow as tf
     import numpy as np

     # Data
     X = np.array([[230.1, 37.8], [44.5, 39.3], [17.2, 45.9], [151.5, 41.3], [180.8, 10.8]])
     y = np.array([22.1, 10.4, 12.0, 16.5, 17.9])

     # Model
     model = tf.keras.Sequential([
         tf.keras.layers.Dense(1, input_dim=2, use_bias=True)
     ])

     model.compile(optimizer='adam', loss='mse')
     model.fit(X, y, epochs=500)

     # Results
     print("Intercept:", model.layers[0].bias.numpy())
     print("Coefficients:", model.layers[0].kernel.numpy().flatten())
     ```

### 2. **PyTorch**
   - **Use case**: Like TensorFlow, PyTorch is another deep learning framework that can be used to implement linear regression, especially if you're comfortable with automatic differentiation and optimization in a more customizable environment.
   - **How to use**: Define a simple neural network with one linear layer and use the `torch.optim` package to optimize the model’s weights.
   - **Example**:
     ```python
     import torch
     import numpy as np

     # Data
     X = torch.tensor([[230.1, 37.8], [44.5, 39.3], [17.2, 45.9], [151.5, 41.3], [180.8, 10.8]], dtype=torch.float32)
     y = torch.tensor([22.1, 10.4, 12.0, 16.5, 17.9], dtype=torch.float32)

     # Model
     model = torch.nn.Linear(2, 1)  # 2 input features, 1 output

     # Loss and optimizer
     criterion = torch.nn.MSELoss()
     optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

     # Training loop
     for epoch in range(1000):
         model.train()
         optimizer.zero_grad()
         outputs = model(X)
         loss = criterion(outputs.squeeze(), y)
         loss.backward()
         optimizer.step()

     # Results
     print("Intercept:", model.bias.item())
     print("Coefficients:", model.weight.detach().numpy().flatten())
     ```

### 3. **MLlib (Apache Spark)**

   - **How to use**: MLlib provides a simple API for linear regression that can handle large datasets across multiple nodes in a cluster.
   - **Example**:

     ```python
     df = spark.createDataFrame(data, ["feature1", "feature2", "label"])
     ```
