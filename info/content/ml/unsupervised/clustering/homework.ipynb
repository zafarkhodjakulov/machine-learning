```python
import pandas as pd
```

```python
df = pd.read_csv('data/Mall_Customers.csv').drop(['CustomerID'], axis=1)
df.columns = ['gender', 'age', 'income', 'spending_score']
```

```python
df.head()
```

```python
df = pd.concat([
    df, pd.get_dummies(df['gender'])
    ], 
    axis=1
).drop(columns=['gender'])
```

```python
df.head()
```

```python
from ipywidgets import Dropdown

x = Dropdown(options=['income', 'age', 'spending_score'])
x
```

```python
y = Dropdown(options=['income', 'age', 'spending_score'])
y
```

```python
from matplotlib import pyplot as plt

df.plot.scatter(x=x.value, y=y.value)
plt.show()
```

```python
from sklearn.cluster import KMeans

k_range = range(1, 10)
see = []

for n in k_range:
    km = KMeans(n_clusters=n)
    km.fit(df)
    see.append(km.inertia_)
```

```python
plt.plot(k_range, see, marker='.')
plt.show()
```

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(df)
```

```python
df.plot.scatter(x=x.value, y=y.value, c=y_pred, colormap='winter', figsize=(8, 6))
plt.show()
```
