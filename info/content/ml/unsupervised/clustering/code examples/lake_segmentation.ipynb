```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from PIL import Image

def show_image(img):
    plt.imshow(img)
    plt.axis(False)
    plt.show()
```

```python
image = np.array(Image.open('lakes.jpg'))
show_image(image)
```

```python
X = image.reshape(-1, 3)
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

pred_img = centroids[labels]
pred_img = pred_img.reshape(image.shape).astype(int)
```

```python
plt.imshow(pred_img)
plt.show()
```
