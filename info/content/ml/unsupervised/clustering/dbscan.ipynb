```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np

X, _ = make_circles(n_samples=500, factor=.5, noise=.03, random_state=4)
Y, _ = make_circles(n_samples=500, factor=.1, noise=.03, random_state=4)
```

```python
Z = np.vstack([X, Y])
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(Z[:,0], Z[:,1], s=5)
plt.show()
```

```python
dbscan = DBSCAN(eps=0.1, min_samples=5)
clusters = dbscan.fit_predict(Z)
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(Z[:,0], Z[:,1], s=5, c=clusters, cmap='viridis')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(Z)

plt.figure(figsize=(4, 3))
plt.scatter(Z[:,0], Z[:,1], s=5, c=clusters, cmap='viridis')
plt.show()
```
